```python
import streamlit as st
from llama_index.readers.file.pdf import SimplePDFReader as PDFReader
from llama_index.vector_stores import ChromaVectorStore
from llama_index.storage.storage_context import StorageContext
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index import VectorStoreIndex
import chromadb

st.title("📚 PDF to ChromaDB Streamlit Example")

# ---------------- Fixed PDF Paths ----------------
pdf_paths = [
    "/content/_OceanofPDF.com_Hands-On_Large_Language_Models_-_Jay_Alammar.pdf",
    "/content/dokumen.pub_practical-natural-language-processing-a-comprehensive-guide-to-building-real-world-nlp-system.pdf",
    "/content/speech and language pocessing.pdf",
]

st.info("📂 Step 1: Starting PDF parsing locally with PDFReader...")
documents = []

for path in pdf_paths:
    try:
        reader = PDFReader()
        docs = reader.load_data(path)
        documents.extend(docs)
        st.success(f"✅ Parsed: {path} ({len(docs)} docs)")
    except Exception as e:
        st.error(f"❌ Failed parsing {path}: {e}")

st.info(f"➡️ Total documents loaded: {len(documents)}")

# ---------------- ChromaDB Connection ----------------
st.info("🗄️ Step 2: Connecting to ChromaDB...")
db = chromadb.Client()
chroma_collection = db.get_or_create_collection("adaptive_education_docs")
st.success("✅ Step 2 Completed: Connected to ChromaDB!")

# ---------------- VectorStore & StorageContext ----------------
st.info("📦 Step 3: Initializing ChromaVectorStore & StorageContext...")
vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
storage_context = StorageContext.from_defaults(vector_store=vector_store)
st.success("✅ Step 3 Completed: VectorStore + StorageContext ready!")

# ---------------- Embedding Model ----------------
st.info("📦 Step 4: Loading local embedding model...")
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-mpnet-base-v2")
st.success("✅ Embedding model loaded successfully!")

# ---------------- Build Vector Store Index ----------------
st.info("⚙️ Step 5: Building the Vector Store Index...")
try:
    index = VectorStoreIndex.from_documents(
        documents,
        storage_context=storage_context,
        embed_model=embed_model
    )
    st.success("✅ Step 5 Completed: Vector Store Index built successfully!")
    st.info(f"➡️ Vectors stored in Chroma: {chroma_collection.count()}")
except Exception as e:
    st.error(f"❌ Error while building index: {e}")
```

```python
import os
import subprocess
import time
import socket
from pyngrok import ngrok
import threading
import sys

# 1. Setup ngrok
print("Setting up ngrok tunnel...")
NGROK_AUTH = "32Qd6m3WC4U3DYjZppZbbWCL3NY_7wSnntQR8UW7mkGeiz2MG"
os.system("ngrok config add-authtoken {}".format(NGROK_AUTH))

# Function to wait until port is ready with better error handling
def wait_for_port(port, host="127.0.0.1", timeout=180):
    start_time = time.time()
    while True:
        try:
            with socket.create_connection((host, port), timeout=2):
                print("Server is running on {}:{}".format(host, port))
                return True
        except OSError as e:
            if time.time() - start_time > timeout:
                print("Server did not start on {}:{} within {} seconds. Last error: {}".format(host, port, timeout, e))
                return False
            elapsed = int(time.time() - start_time)
            print("Waiting for server... ({}s elapsed)".format(elapsed))
            time.sleep(3)

# Function to monitor process output
def monitor_streamlit_output(process):
    try:
        for line in iter(process.stdout.readline, ''):
            if line:
                print("Streamlit: {}".format(line.strip()))
                if "Network URL" in line or "External URL" in line or "Local URL" in line:
                    print("Streamlit server started successfully!")
    except Exception as e:
        print("Monitor error: {}".format(e))

# 2. First, fix dependency issues by uninstalling problematic packages
print("🔧 Fixing dependency conflicts...")
os.system("pip uninstall huggingface_hub sentence-transformers -y --quiet")
time.sleep(2)

# 3. Install compatible versions in correct order
print("📦 Installing compatible packages in order...")
dependency_commands = [
    "pip install huggingface_hub==0.16.4 --quiet --no-cache-dir",
    "pip install tokenizers==0.13.3 --quiet --no-cache-dir",
    "pip install transformers==4.30.0 --quiet --no-cache-dir",
    "pip install sentence-transformers==2.2.2 --quiet --no-cache-dir"
]

for cmd in dependency_commands:
    print(f"Running: {cmd}")
    os.system(cmd)
    time.sleep(1)

# 4. Create ROBUST RAG app.py with fallback embedding systems
app_code = '''
import os
import streamlit as st
from gtts import gTTS
import whisper
from st_audiorec import st_audiorec
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, AutoModel
import time
import io
from datetime import datetime
import torch
import torch.nn.functional as F
import gc
from threading import Lock
import warnings
import chromadb
import numpy as np
import json
warnings.filterwarnings("ignore")

# Page Configuration
st.set_page_config(
    page_title="MR NLP RAG Chatbot",
    page_icon="📚",
    layout="wide",
    initial_sidebar_state="expanded"
)

# Custom CSS for styling
st.markdown("""
<style>
    .main-container {
        background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
        padding: 20px;
        border-radius: 15px;
        margin: 10px 0;
        color: white;
    }
    .response-container {
        background: #f8f9fa;
        padding: 20px;
        border-radius: 10px;
        margin: 15px 0;
        border-left: 5px solid #667eea;
    }
    .rag-context {
        background: #e8f4fd;
        padding: 15px;
        border-radius: 8px;
        margin: 10px 0;
        border-left: 4px solid #2196F3;
        font-size: 0.9em;
    }
    .source-box {
        background: #fff3cd;
        padding: 10px;
        border-radius: 5px;
        margin: 5px 0;
        border-left: 3px solid #ffc107;
        font-size: 0.85em;
    }
    .stButton > button {
        background: linear-gradient(90deg, #667eea 0%, #764ba2 100%);
        color: white;
        border: none;
        border-radius: 25px;
        padding: 10px 30px;
        font-weight: bold;
        width: 100%;
    }
    .loading-text {
        color: #667eea;
        font-weight: bold;
        text-align: center;
    }
</style>
""", unsafe_allow_html=True)

# Header
st.title("📚 MR NLP Robust RAG Chatbot")
st.subheader("Qwen + Robust ChromaDB Integration with Fallback Systems")
st.markdown("---")

# Sidebar Configuration
with st.sidebar:
    st.header("⚙️ Configuration")

    # Model Settings
    st.subheader("🧠 Qwen Model Settings")
    model_temp = st.slider("Temperature", 0.1, 2.0, 0.8, 0.1)
    max_tokens = st.slider("Max New Tokens", 32, 128, 64, 16)
    top_p = st.slider("Top P", 0.1, 1.0, 0.9, 0.1)

    # RAG Settings
    st.subheader("📚 RAG Settings")
    use_rag = st.checkbox("Enable RAG (Knowledge Base)", value=True, help="Use vector database for context")
    rag_similarity_top_k = st.slider("Retrieved Documents", 1, 10, 3, 1, help="Number of relevant documents to retrieve")
    rag_similarity_threshold = st.slider("Similarity Threshold", 0.1, 1.0, 0.7, 0.1, help="Minimum relevance score")
    show_sources = st.checkbox("Show Source Documents", value=True, help="Display retrieved document chunks")
    max_context_length = st.slider("Max Context Length", 500, 2000, 1000, 100, help="Maximum characters from retrieved docs")

    # Embedding Method Selection
    st.subheader("🔤 Embedding Method")
    embedding_method = st.selectbox("Choose Embedding System:",
        ["Auto (Try SentenceTransformers first)", "Transformers Fallback", "Simple TF-IDF"],
        help="Different embedding systems for compatibility")

    # Performance Settings
    st.subheader("⚡ Performance")
    use_4bit = st.checkbox("Use 4-bit Quantization", value=True, help="Reduces memory usage")
    use_cpu_only = st.checkbox("Force CPU Only", value=False, help="Use if GPU causes issues")
    enable_cache = st.checkbox("Enable Response Cache", value=True, help="Cache responses for speed")

    # Audio Settings
    st.subheader("🎵 Audio Settings")
    tts_speed = st.selectbox("TTS Speed", ["slow", "normal", "fast"], index=1)
    audio_lang = st.selectbox("Language", ["en", "ar", "fr", "es"], index=0)
    enable_tts = st.checkbox("Enable Text-to-Speech", value=True)
    enable_voice = st.checkbox("Enable Voice Input", value=True)

# Initialize session state
if 'initialized' not in st.session_state:
    st.session_state.initialized = False
    st.session_state.model_loaded = False
    st.session_state.whisper_loaded = False
    st.session_state.rag_loaded = False
    st.session_state.conversation_history = []
    st.session_state.response_cache = {}
    st.session_state.model_lock = Lock()
    st.session_state.loading_complete = False
    st.session_state.embedding_system = None

# Multiple embedding system fallbacks
class RobustEmbeddingSystem:
    def __init__(self):
        self.model = None
        self.tokenizer = None
        self.system_type = None

    def load_sentence_transformers(self):
        """Try to load SentenceTransformers"""
        try:
            from sentence_transformers import SentenceTransformer
            self.model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
            self.system_type = "SentenceTransformers"
            return True
        except Exception as e:
            st.warning(f"SentenceTransformers failed: {e}")
            return False

    def load_transformers_fallback(self):
        """Fallback to direct transformers with mean pooling"""
        try:
            model_name = "sentence-transformers/all-mpnet-base-v2"
            self.tokenizer = AutoTokenizer.from_pretrained(model_name)
            self.model = AutoModel.from_pretrained(model_name)
            self.model.eval()
            self.system_type = "Transformers"
            return True
        except Exception as e:
            st.warning(f"Transformers fallback failed: {e}")
            return False

    def load_tfidf_fallback(self):
        """Simple TF-IDF fallback"""
        try:
            from sklearn.feature_extraction.text import TfidfVectorizer
            from sklearn.metrics.pairwise import cosine_similarity
            self.model = TfidfVectorizer(max_features=5000, stop_words='english')
            self.system_type = "TF-IDF"
            self.fitted = False
            return True
        except Exception as e:
            st.warning(f"TF-IDF fallback failed: {e}")
            return False

    def mean_pooling(self, model_output, attention_mask):
        """Mean pooling for transformer embeddings"""
        token_embeddings = model_output[0]
        input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
        return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)

    def encode(self, texts):
        """Encode texts using the loaded system"""
        if isinstance(texts, str):
            texts = [texts]

        if self.system_type == "SentenceTransformers":
            return self.model.encode(texts)

        elif self.system_type == "Transformers":
            with torch.no_grad():
                encoded = self.tokenizer(texts, padding=True, truncation=True, return_tensors='pt', max_length=512)
                model_output = self.model(**encoded)
                embeddings = self.mean_pooling(model_output, encoded['attention_mask'])
                embeddings = F.normalize(embeddings, p=2, dim=1)
                return embeddings.numpy()

        elif self.system_type == "TF-IDF":
            if not hasattr(self, 'fitted') or not self.fitted:
                # This is just for query encoding; collection fitting happens separately
                pass
            return self.model.transform(texts).toarray()

        else:
            raise ValueError("No embedding system loaded")

@st.cache_resource(show_spinner=False)
def initialize_robust_rag_system():
    """Initialize ChromaDB and robust embedding system"""
    try:
        # Initialize ChromaDB client
        chroma_client = chromadb.Client()

        # Get or create collection
        collection_name = "adaptive_education_docs"
        try:
            collection = chroma_client.get_collection(name=collection_name)
            st.success(f"✅ Found existing collection: {collection_name}")
        except Exception:
            st.warning(f"⚠️ Collection '{collection_name}' not found. Creating new one...")
            collection = chroma_client.create_collection(name=collection_name)

        # Initialize robust embedding system
        embedding_system = RobustEmbeddingSystem()

        # Try loading embedding systems in order of preference
        if embedding_method == "Auto (Try SentenceTransformers first)":
            if not embedding_system.load_sentence_transformers():
                if not embedding_system.load_transformers_fallback():
                    embedding_system.load_tfidf_fallback()
        elif embedding_method == "Transformers Fallback":
            if not embedding_system.load_transformers_fallback():
                embedding_system.load_tfidf_fallback()
        else:  # TF-IDF
            embedding_system.load_tfidf_fallback()

        # Check collection size
        collection_count = collection.count()

        return collection, embedding_system, collection_count

    except Exception as e:
        st.error(f"❌ RAG initialization failed: {e}")
        return None, None, 0

# Optimized Qwen model initialization
@st.cache_resource(show_spinner=False)
def initialize_qwen_model():
    """Initialize Qwen model with optimizations"""
    try:
        model_id = "Qwen/Qwen1.5-1.8B"

        # Determine device
        if use_cpu_only:
            device = "cpu"
            torch_dtype = torch.float32
        else:
            device = "cuda" if torch.cuda.is_available() else "cpu"
            torch_dtype = torch.float16 if device == "cuda" else torch.float32

        # Quantization config for memory efficiency
        quantization_config = None
        if use_4bit and not use_cpu_only and torch.cuda.is_available():
            try:
                quantization_config = BitsAndBytesConfig(
                    load_in_4bit=True,
                    bnb_4bit_compute_dtype=torch.float16,
                    bnb_4bit_use_double_quant=True,
                    bnb_4bit_quant_type="nf4"
                )
            except Exception as e:
                st.warning(f"4-bit quantization not available: {e}")
                quantization_config = None

        # Load tokenizer
        tokenizer = AutoTokenizer.from_pretrained(
            model_id,
            trust_remote_code=True,
            use_fast=True
        )

        if tokenizer.pad_token is None:
            tokenizer.pad_token = tokenizer.eos_token

        # Load model with optimizations
        model_kwargs = {
            "torch_dtype": torch_dtype,
            "trust_remote_code": True,
            "low_cpu_mem_usage": True,
            "device_map": "auto" if not use_cpu_only and torch.cuda.is_available() else None
        }

        if quantization_config:
            model_kwargs["quantization_config"] = quantization_config

        model = AutoModelForCausalLM.from_pretrained(model_id, **model_kwargs)

        if use_cpu_only or not torch.cuda.is_available():
            model = model.to("cpu")

        # Set to eval mode for inference
        model.eval()

        return model, tokenizer, device

    except Exception as e:
        st.error(f"Error loading Qwen model: {e}")
        return None, None, None

@st.cache_resource(show_spinner=False)
def load_whisper_model():
    """Load Whisper model for voice recognition"""
    try:
        if not enable_voice:
            return None
        return whisper.load_model("tiny")  # Use tiny model for speed
    except Exception as e:
        st.error(f"Error loading Whisper: {e}")
        return None

# Progressive loading with status
def initialize_components():
    if not st.session_state.initialized:

        # Create loading status
        loading_placeholder = st.empty()
        progress_bar = st.progress(0)

        try:
            # Load RAG system first
            if use_rag:
                loading_placeholder.markdown('<p class="loading-text">📚 Initializing Robust RAG System...</p>', unsafe_allow_html=True)
                progress_bar.progress(10)

                collection, embedding_system, collection_count = initialize_robust_rag_system()
                if collection and embedding_system:
                    st.session_state.chroma_collection = collection
                    st.session_state.embedding_system = embedding_system
                    st.session_state.collection_count = collection_count
                    st.session_state.rag_loaded = True
                    progress_bar.progress(30)
                    st.info(f"✅ Using {embedding_system.system_type} embedding system")
                else:
                    st.session_state.rag_loaded = False

            # Load Qwen model
            loading_placeholder.markdown('<p class="loading-text">🧠 Loading Qwen 1.5-1.8B Model...</p>', unsafe_allow_html=True)
            progress_bar.progress(50)

            model, tokenizer, device = initialize_qwen_model()

            if model and tokenizer:
                st.session_state.model = model
                st.session_state.tokenizer = tokenizer
                st.session_state.device = device
                st.session_state.model_loaded = True
                progress_bar.progress(80)
            else:
                raise Exception("Failed to load Qwen model")

            # Load whisper if needed
            if enable_voice:
                loading_placeholder.markdown('<p class="loading-text">🎤 Loading Voice Recognition...</p>', unsafe_allow_html=True)
                progress_bar.progress(90)

                whisper_model = load_whisper_model()
                if whisper_model:
                    st.session_state.whisper_model = whisper_model
                    st.session_state.whisper_loaded = True

            progress_bar.progress(100)
            loading_placeholder.markdown('<p class="loading-text">✅ All Components Loaded Successfully!</p>', unsafe_allow_html=True)

            st.session_state.initialized = True
            st.session_state.loading_complete = True

            # Clear loading indicators after 2 seconds
            time.sleep(2)
            loading_placeholder.empty()
            progress_bar.empty()

        except Exception as e:
            loading_placeholder.empty()
            progress_bar.empty()
            st.error(f"❌ Initialization failed: {e}")
            st.info("🔄 Try refreshing the page or switching embedding methods.")
            return

# Initialize on first run
if not st.session_state.loading_complete:
    initialize_components()

# Status display
if st.session_state.loading_complete:
    col1, col2, col3, col4 = st.columns(4)
    with col1:
        if st.session_state.model_loaded:
            st.success("✅ Qwen Model Ready")
        else:
            st.error("❌ Qwen Model Failed")

    with col2:
        if use_rag and st.session_state.rag_loaded:
            vector_count = st.session_state.collection_count
            embedding_type = st.session_state.embedding_system.system_type
            st.success(f"✅ RAG Ready ({vector_count} docs)")
            st.caption(f"Using: {embedding_type}")
        elif not use_rag:
            st.info("📚 RAG Disabled")
        else:
            st.error("❌ RAG Failed")

    with col3:
        if enable_voice and st.session_state.whisper_loaded:
            st.success("✅ Voice Ready")
        elif not enable_voice:
            st.info("🔇 Voice Disabled")
        else:
            st.error("❌ Voice Failed")

    with col4:
        device_info = st.session_state.get('device', 'unknown')
        if use_4bit and device_info == 'cuda':
            st.info("💾 4-bit GPU")
        elif device_info == 'cuda':
            st.info("🚀 GPU Mode")
        else:
            st.info("💾 CPU Mode")

# Only proceed if model is loaded
if not st.session_state.model_loaded:
    if st.session_state.loading_complete:
        st.error("🚨 Critical Error: Qwen model failed to load. Please refresh the page.")
        if st.button("🔄 Retry Loading"):
            st.cache_resource.clear()
            st.rerun()
    st.stop()

if st.session_state.loading_complete:
    if use_rag and st.session_state.rag_loaded:
        st.success("🎉 Robust RAG-Enhanced Qwen is ready! Ask questions about your documents.")
    else:
        st.success("🎉 Qwen is ready! (RAG disabled or unavailable)")

# Chat Interface
st.markdown("### 💬 Chat Interface")

# Input method selection
mode = st.radio("Choose input method:", ["💬 Typing", "🎤 Voice"], horizontal=True)

user_q = None

if mode == "💬 Typing":
    user_q = st.text_input(
        "Ask your question:",
        placeholder="What would you like to know from your documents?",
        key="text_input"
    )

elif mode == "🎤 Voice" and enable_voice and st.session_state.whisper_loaded:
    st.write("**🎤 Record your question:**")
    wav_audio_data = st_audiorec(key="audio_recorder")

    if wav_audio_data:
        with st.spinner("🎯 Transcribing audio..."):
            try:
                # Save audio file
                audio_path = f"temp_audio_{int(time.time())}.wav"
                with open(audio_path, "wb") as f:
                    f.write(wav_audio_data)

                # Transcribe
                result = st.session_state.whisper_model.transcribe(
                    audio_path,
                    language=audio_lang if audio_lang != 'en' else None
                )
                user_q = result.get("text", "").strip()

                if user_q:
                    st.success(f"✅ Transcribed: {user_q}")

                # Cleanup
                if os.path.exists(audio_path):
                    os.remove(audio_path)

            except Exception as e:
                st.error(f"❌ Transcription failed: {e}")

elif mode == "🎤 Voice":
    st.warning("🔇 Voice input is disabled or failed to load.")

# Robust RAG retrieval function
def retrieve_relevant_context(query, top_k=3):
    """Retrieve relevant context using robust embedding system"""
    try:
        if not (use_rag and st.session_state.rag_loaded):
            return "", []

        embedding_system = st.session_state.embedding_system

        # Generate query embedding based on system type
        if embedding_system.system_type in ["SentenceTransformers", "Transformers"]:
            # Use semantic embeddings
            query_embedding = embedding_system.encode([query])

            # Search in ChromaDB
            results = st.session_state.chroma_collection.query(
                query_embeddings=query_embedding.tolist(),
                n_results=top_k,
                include=["documents", "metadatas", "distances"]
            )

            # Process results
            relevant_docs = []
            context_parts = []

            if results and results.get("documents") and len(results["documents"]) > 0:
                documents = results["documents"][0]  # First query results
                distances = results.get("distances", [[]])[0]
                metadatas = results.get("metadatas", [[]])[0]

                for i, (doc, distance, metadata) in enumerate(zip(documents, distances, metadatas)):
                    # Convert distance to similarity score
                    similarity = 1 - distance

                    # Filter by similarity threshold
                    if similarity >= rag_similarity_threshold:
                        # Limit document length
                        doc_text = doc[:max_context_length] if len(doc) > max_context_length else doc
                        context_parts.append(f"Document {i+1}: {doc_text}")

                        relevant_docs.append({
                            "content": doc[:300] + "..." if len(doc) > 300 else doc,
                            "full_content": doc,
                            "score": similarity,
                            "metadata": metadata or {}
                        })

        elif embedding_system.system_type == "TF-IDF":
            # For TF-IDF, we need to get all documents first and then search
            all_results = st.session_state.chroma_collection.get(include=["documents", "metadatas"])

            if all_results and all_results.get("documents"):
                all_docs = all_results["documents"]
                all_metadatas = all_results.get("metadatas", [{}] * len(all_docs))

                # Fit TF-IDF on all documents + query
                corpus = all_docs + [query]
                tfidf_matrix = embedding_system.model.fit_transform(corpus)
                embedding_system.fitted = True

                # Calculate similarities
                query_vec = tfidf_matrix[-1]  # Last item is the query
                doc_vecs = tfidf_matrix[:-1]  # All but last are documents

                from sklearn.metrics.pairwise import cosine_similarity
                similarities = cosine_similarity(query_vec, doc_vecs).flatten()

                # Get top-k most similar documents
                top_indices = similarities.argsort()[-top_k:][::-1]

                relevant_docs = []
                context_parts = []

                for i, idx in enumerate(top_indices):
                    similarity = similarities[idx]

                    if similarity >= rag_similarity_threshold:
                        doc = all_docs[idx]
                        metadata = all_metadatas[idx] if idx < len(all_metadatas) else {}

                        # Limit document length
                        doc_text = doc[:max_context_length] if len(doc) > max_context_length else doc
                        context_parts.append(f"Document {i+1}: {doc_text}")

                        relevant_docs.append({
                            "content": doc[:300] + "..." if len(doc) > 300 else doc,
                            "full_content": doc,
                            "score": similarity,
                            "metadata": metadata
                        })
            else:
                return "", []

        context = "\\n\\n".join(context_parts)
        return context, relevant_docs

    except Exception as e:
        st.warning(f"RAG retrieval failed: {e}")
        return "", []

# Enhanced Qwen generation with RAG
def generate_rag_enhanced_response(prompt, user_question, rag_context=""):
    """Generate response using Qwen model with RAG context"""
    try:
        with st.session_state.model_lock:
            # Check cache first
            cache_key = f"{user_question}_{model_temp}_{max_tokens}_{len(rag_context)}"
            if enable_cache and cache_key in st.session_state.response_cache:
                return st.session_state.response_cache[cache_key]

            # Create enhanced prompt with RAG context
            if rag_context:
                enhanced_prompt = f"""Based on the following relevant information from the documents, please answer the question:

Relevant Information:
{rag_context}

Question: {user_question}

Please provide a comprehensive answer based on the information above. If the information doesn't fully answer the question, acknowledge what you can answer and what might need additional information.

Answer:"""
            else:
                enhanced_prompt = prompt

            # Tokenize input
            inputs = st.session_state.tokenizer(
                enhanced_prompt,
                return_tensors="pt",
                truncation=True,
                max_length=1024,  # Increased for RAG context
                padding=True
            )

            # Move to device
            inputs = {k: v.to(st.session_state.device) for k, v in inputs.items()}

            # Generate with optimized parameters
            with torch.no_grad():
                outputs = st.session_state.model.generate(
                    **inputs,
                    max_new_tokens=max_tokens,
                    temperature=model_temp,
                    top_p=top_p,
                    do_sample=True,
                    pad_token_id=st.session_state.tokenizer.eos_token_id,
                    eos_token_id=st.session_state.tokenizer.eos_token_id,
                    repetition_penalty=1.1,
                    num_return_sequences=1,
                    use_cache=True
                )

            # Decode response
            response = st.session_state.tokenizer.decode(
                outputs[0][inputs['input_ids'].shape[1]:],
                skip_special_tokens=True
            ).strip()

            # Cache the response
            if enable_cache:
                st.session_state.response_cache[cache_key] = response

                # Limit cache size
                if len(st.session_state.response_cache) > 50:
                    keys_to_remove = list(st.session_state.response_cache.keys())[:10]
                    for key in keys_to_remove:
                        del st.session_state.response_cache[key]

            return response

    except Exception as e:
        raise Exception(f"RAG-enhanced generation failed: {e}")

# Query Processing with Robust RAG
if user_q and user_q.strip():
    if st.button("🚀 Ask Robust RAG Qwen", use_container_width=True):
        start_time = time.time()

        # Show immediate feedback
        response_placeholder = st.empty()
        progress_placeholder = st.empty()

        try:
            user_q = user_q.strip()

            # Show progress
            with response_placeholder.container():
                st.markdown("🔍 **Searching knowledge base...**")
                progress = progress_placeholder.progress(0)

            # Step 1: Retrieve relevant context
            rag_context = ""
            sources = []

            if use_rag and st.session_state.rag_loaded:
                progress.progress(20)
                rag_context, sources = retrieve_relevant_context(user_q, rag_similarity_top_k)
                if rag_context:
                    progress.progress(40)
                    with response_placeholder.container():
                        st.markdown("📚 **Found relevant information! Generating response...**")
                else:
                    with response_placeholder.container():
                        st.markdown("⚠️ **No highly relevant documents found. Using general knowledge...**")

            progress.progress(50)

            # Add to conversation history
            st.session_state.conversation_history.append(f"User: {user_q}")

            # Create context-aware prompt
            conversation_context = ""
            if len(st.session_state.conversation_history) > 1:
                recent = st.session_state.conversation_history[-6:]
                conversation_context = "\\n".join(recent) + "\\n"

            # Create system prompt
            system_prompt = "You are a helpful AI assistant with access to relevant documents. Provide clear, accurate responses based on the available information."

            if conversation_context:
                prompt = f"{system_prompt}\\n\\nConversation history:\\n{conversation_context}\\nUser: {user_q}\\nAssistant:"
            else:
                prompt = f"{system_prompt}\\n\\nUser: {user_q}\\nAssistant:"

            progress.progress(60)

            # Generate response
            try:
                final_answer = generate_rag_enhanced_response(prompt, user_q, rag_context)

                if not final_answer or len(final_answer.strip()) < 5:
                    if rag_context:
                        final_answer = f"Based on the documents, I found some relevant information but couldn't generate a complete response. Please try rephrasing your question about '{user_q}'."
                    else:
                        final_answer = f"I understand your question about '{user_q}'. Could you please provide more details or rephrase your question?"

            except Exception as model_error:
                st.warning(f"Generation issue: {model_error}")
                final_answer = f"I encountered a technical issue while processing your question. Please try again."

            progress.progress(80)

            # Add AI response to history
            st.session_state.conversation_history.append(f"Assistant: {final_answer}")

            # Keep history manageable
            if len(st.session_state.conversation_history) > 12:
                st.session_state.conversation_history = st.session_state.conversation_history[-8:]

            processing_time = time.time() - start_time
            progress.progress(100)

            # Clear progress indicators
            response_placeholder.empty()
            progress_placeholder.empty()

            # Display Results
            st.markdown("---")

            # Show RAG context if available
            if rag_context and sources and show_sources:
                st.subheader("📚 Retrieved Knowledge")
                for i, source in enumerate(sources):
                    with st.expander(f"📄 Source {i+1} (Score: {source['score']:.3f})", expanded=False):
                        st.markdown(f'<div class="source-box">{source["content"]}</div>', unsafe_allow_html=True)
                        if source['metadata']:
                            st.caption(f"Metadata: {source['metadata']}")
                        if st.button(f"Show Full Content {i+1}", key=f"full_{i}"):
                            st.text_area(f"Full Document {i+1}:", source['full_content'], height=200)
                st.markdown("---")

            # Main response
            st.subheader("🤖 Robust RAG Response")
            st.markdown(f'<div class="response-container">{final_answer}</div>', unsafe_allow_html=True)

            # Generate and play audio
            if enable_tts:
                try:
                    with st.spinner("🔊 Generating speech..."):
                        tts_text = final_answer[:400] + "..." if len(final_answer) > 400 else final_answer
                        tts = gTTS(text=tts_text, lang=audio_lang, slow=(tts_speed=="slow"))
                        audio_file = f"robust_rag_response_{int(time.time())}.mp3"
                        tts.save(audio_file)

                    st.audio(audio_file, format="audio/mp3")

                    def cleanup_audio():
                        time.sleep(30)
                        if os.path.exists(audio_file):
                            os.remove(audio_file)

                    threading.Thread(target=cleanup_audio, daemon=True).start()

                except Exception as tts_error:
                    st.warning(f"🔇 TTS failed: {tts_error}")

            # Response metadata
            col1, col2, col3 = st.columns(3)

            with col1:
                st.info(f"⏱️ **Processing Time:** {processing_time:.2f}s")
                embedding_type = st.session_state.embedding_system.system_type if st.session_state.rag_loaded else "None"
                st.info(f"📊 **Embedding:** {embedding_type}")

            with col2:
                st.info(f"📝 **Response Length:** {len(final_answer)} chars")
                st.info(f"📚 **Sources Used:** {len(sources)}")

            with col3:
                st.info(f"🕒 **Timestamp:** {datetime.now().strftime('%H:%M:%S')}")
                st.info(f"🎯 **RAG Mode:** {'Enabled' if rag_context else 'Disabled'}")

            # RAG settings display
            if use_rag and st.session_state.rag_loaded:
                with st.expander("🔧 RAG Settings", expanded=False):
                    st.write(f"**Embedding System:** {st.session_state.embedding_system.system_type}")
                    st.write(f"**Similarity Top K:** {rag_similarity_top_k}")
                    st.write(f"**Similarity Threshold:** {rag_similarity_threshold}")
                    st.write(f"**Max Context Length:** {max_context_length}")
                    st.write(f"**Vector Database Size:** {st.session_state.collection_count} documents")
                    st.write(f"**Sources Retrieved:** {len(sources)}")
                    st.write(f"**Context Length:** {len(rag_context)} characters")

            # Conversation history
            if st.session_state.conversation_history:
                with st.expander("📜 Conversation History", expanded=False):
                    for msg in st.session_state.conversation_history[-8:]:
                        st.text(msg)

            # Feedback section
            st.markdown("---")
            col1, col2, col3 = st.columns(3)
            with col1:
                if st.button("👍 Excellent Answer", key="excellent"):
                    st.success("Thank you! Robust RAG system is working well!")
            with col2:
                if st.button("👎 Needs Improvement", key="needs_work"):
                    st.info("Thanks! Try switching embedding methods or adjusting settings.")
            with col3:
                if st.button("🔄 Try Again", key="retry"):
                    st.rerun()

        except Exception as e:
            response_placeholder.empty()
            progress_placeholder.empty()
            st.error(f"❌ An error occurred: {e}")
            st.info("💡 Try: switching embedding method, adjusting RAG settings, or refreshing the page.")

            # Force cleanup
            gc.collect()
            if torch.cuda.is_available():
                torch.cuda.empty_cache()

# Utility buttons
st.markdown("---")
col1, col2, col3, col4 = st.columns(4)

with col1:
    if st.button("🗑️ Clear History", type="secondary"):
        st.session_state.conversation_history = []
        st.success("History cleared!")

with col2:
    if st.button("🧹 Clear Cache", type="secondary"):
        st.session_state.response_cache = {}
        st.success("Cache cleared!")

with col3:
    if st.button("♻️ Reset Session", type="secondary"):
        # Keep essential components
        keys_to_keep = ['model', 'tokenizer', 'device', 'whisper_model', 'chroma_collection', 'embedding_system', 'collection_count', 'initialized', 'model_loaded', 'whisper_loaded', 'rag_loaded', 'loading_complete', 'model_lock']
        for key in list(st.session_state.keys()):
            if key not in keys_to_keep:
                del st.session_state[key]
        st.session_state.conversation_history = []
        st.session_state.response_cache = {}
        st.success("Session reset!")

with col4:
    if st.button("🔄 Restart App", type="secondary"):
        st.cache_resource.clear()
        st.rerun()

# Memory management and stats
try:
    if torch.cuda.is_available() and not use_cpu_only:
        gpu_memory = torch.cuda.memory_allocated() / 1024**3  # GB
        gpu_cached = torch.cuda.memory_reserved() / 1024**3   # GB
        memory_info = f"GPU: {gpu_memory:.2f}GB used, {gpu_cached:.2f}GB cached"
    else:
        memory_info = "Running on CPU"

    cache_info = f"Response cache: {len(st.session_state.response_cache)} entries"
    device_info = f"Device: {st.session_state.get('device', 'unknown')}"

    if use_rag and st.session_state.rag_loaded:
        vector_info = f"Vector DB: {st.session_state.collection_count} documents"
        embedding_info = f"Embedding: {st.session_state.embedding_system.system_type}"
        st.caption(f"💾 {memory_info} | 📋 {cache_info} | 📚 {vector_info} | 🔤 {embedding_info} | 🖥️ {device_info}")
    else:
        st.caption(f"💾 {memory_info} | 📋 {cache_info} | 🖥️ {device_info}")
except:
    st.caption("🤖 MR NLP Robust RAG Chatbot - Multiple Embedding System Fallbacks")

# Footer
st.markdown("---")
st.markdown("**📚 MR NLP Robust RAG Chatbot** • Multi-System Embedding • Qwen + ChromaDB")
st.caption("🧠 Qwen 1.5-1.8B • 🔤 SentenceTransformers/Transformers/TF-IDF Fallbacks • ⚡ Optimized Performance • 🎤 Voice Enabled")
'''

# Write the robust RAG app.py file
with open("app.py", "w", encoding="utf-8") as f:
    f.write(app_code)

print("✅ Created ROBUST RAG-ENHANCED app.py with multiple embedding system fallbacks")

# 5. Install remaining basic packages
basic_packages = [
    "streamlit==1.28.0",
    "gtts==2.3.2",
    "openai-whisper==20231117",
    "streamlit-audiorec==0.1.0",
    "torch==2.1.0",
    "torchaudio==2.1.0",
    "accelerate==0.25.0",
    "bitsandbytes==0.41.3",
    "optimum==1.16.0",
    "soundfile==0.12.1",
    "chromadb==0.4.15",
    "numpy==1.24.3",
    "scikit-learn==1.3.0"  # For TF-IDF fallback
]

for package in basic_packages:
    try:
        print(f"Installing {package}...")
        os.system(f"pip install {package} --quiet --no-cache-dir")
    except Exception as e:
        print(f"Warning: Failed to install {package}: {e}")

print("✅ Basic packages installed")

# 6. Kill existing processes
print("🧹 Cleaning up existing processes...")
try:
    os.system("pkill -f streamlit")
    os.system("pkill -f ngrok")
    time.sleep(3)
except:
    pass

# 7. Run robust RAG-enhanced Streamlit app
print("🚀 Launching ROBUST RAG-Enhanced Chatbot...")

cmd = [
    sys.executable, "-m", "streamlit", "run", "app.py",
    "--server.port", "8501",
    "--server.address", "0.0.0.0",
    "--server.headless", "true",
    "--server.enableCORS", "false",
    "--server.enableXsrfProtection", "false",
    "--server.maxUploadSize", "200",
    "--server.maxMessageSize", "200",
    "--browser.gatherUsageStats", "false",
    "--global.developmentMode", "false"
]

process = subprocess.Popen(
    cmd,
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    universal_newlines=True,
    bufsize=1
)

# Start monitoring
monitor_thread = threading.Thread(target=monitor_streamlit_output, args=(process,))
monitor_thread.daemon = True
monitor_thread.start()

# 8. Wait for server
print("⏳ Waiting for robust RAG server (2-4 minutes for full initialization)...")
server_ready = wait_for_port(8501, timeout=300)

if not server_ready:
    print("❌ Failed to start robust RAG server")
    if process.poll() is not None:
        try:
            output = process.stdout.read()
            print(f"Error output: {output}")
        except:
            pass
    sys.exit(1)

# 9. Setup ngrok
print("🌐 Setting up ngrok tunnel...")
try:
    ngrok.kill()
    time.sleep(2)

    public_url = ngrok.connect(8501, proto="http")
    print(f"✅ SUCCESS! Robust RAG Chatbot is live at: {public_url}")

    print("\n" + "="*100)
    print("📚 ROBUST RAG-ENHANCED CHATBOT FEATURES:")
    print("   🧠 Qwen 1.5-1.8B language model")
    print("   🔤 Multiple embedding system fallbacks:")
    print("      1️⃣ SentenceTransformers (primary)")
    print("      2️⃣ Direct Transformers with mean pooling (fallback)")
    print("      3️⃣ TF-IDF vectorization (final fallback)")
    print("   📚 Direct ChromaDB integration")
    print("   🔍 Robust similarity search with configurable thresholds")
    print("   📄 Source document attribution with scores")
    print("   ⚙️ Real-time embedding method switching")
    print("   💾 Smart caching and memory management")
    print("   🎤 Voice input/output capabilities")
    print("="*100)

    print(f"\n📚 ROBUST SYSTEM ADVANTAGES:")
    print("   ✅ Handles HuggingFace Hub version conflicts automatically")
    print("   🔄 Falls back to working embedding method if one fails")
    print("   📊 Shows which embedding system is active")
    print("   🛠️ User can manually select embedding method in sidebar")
    print("   ⚡ TF-IDF fallback works even without neural embeddings")
    print("   🔧 Compatible with older and newer package versions")

    print(f"\n🎯 USAGE TIPS:")
    print("   🔥 First response: 60-120s (model loading + embedding system selection)")
    print("   ⚡ Subsequent responses: 5-20s depending on embedding method")
    print("   📚 Ask questions about your PDF content (NLP, LLMs, Speech Processing)")
    print("   ⚙️ Switch embedding method in sidebar if one isn't working well")
    print("   🔍 Adjust similarity threshold for more/fewer relevant documents")
    print("   💡 TF-IDF method works well for keyword-based searches")
    print("   🧠 SentenceTransformers/Transformers better for semantic similarity")

    print("\n🔄 App is running. Press Ctrl+C to stop...")
    try:
        while True:
            time.sleep(20)
            if process.poll() is not None:
                print("❌ Process terminated!")
                break
            tunnels = ngrok.get_tunnels()
            if not tunnels:
                print("🔄 Reconnecting ngrok...")
                public_url = ngrok.connect(8501, proto="http")
                print(f"✅ Reconnected: {public_url}")
    except KeyboardInterrupt:
        print("\n🛑 Shutting down...")

except Exception as e:
    print(f"❌ Ngrok failed: {e}")

finally:
    print("🧹 Cleaning up...")
    try:
        ngrok.kill()
        if process and process.poll() is None:
            process.terminate()
            process.wait(timeout=5)
    except:
        pass
    print("✅ Cleanup complete")
```
